# Expect the unexpected

```python
import random as r
c = 0
def printer(i):
    global c
    c+=1
    if i <10:
        for j in range(0,i):
            print(j,end=', ')
            printer(j)
        print()
    print("I died :( , I was:  ",c)

i = r.randint(0,1000)
for j in range(0,i):
    printer(j)
```

```python
with open('mbox.txt','r') as file:
    f = open("Written1.txt","w")
    l = file.readline()
    while l:
        if l.startswith('From'):
            k=l.split()
            f.write(k[1]+"\n")
        l = file.readline()
```

```python
emails = dict()

with open('Written1.txt','r') as file:
    l=file.readline().strip()
    while l:
        if l in emails.keys():
            emails[l]+=1
        else:
            emails[l]=1
        l = file.readline().strip()
for i,j in emails.items():
    print(f'email: {i} f: {j}')
print("\nThe num of unique emails is: ",len(emails))
```
